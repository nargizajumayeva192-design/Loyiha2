# **2-bo'lim: Dispersiya o'lchovlari (Range, Variance, Standard Deviation)**


## **1-masala:** Ikki doâ€˜konning 7 kunlik savdo tushumi (mln soâ€˜m)dan foydalanib barqarorlikni aniqlash (oâ€˜rta â†’ qiyin).

## **Topshiriqlar:**

### 1.   Har ikkala doâ€˜kon uchun Range ni hisoblang?
### 2.   Variance va Standard Deviation ni toping?
### 3.   Qaysi doâ€˜kon savdosi barqarorroq?
### 4.   Faqat dispersiya asosida xulosa yozing





```python
# Shu ma'lumotlar bilan ishlang.
import statistics as stat
shop_A = [12.5, 12.8, 13.0, 12.9, 13.1, 12.7, 12.9]
shop_B = [10.0, 14.5, 13.8, 11.2, 15.0, 9.5, 16.0]
mean_A = round(stat.mean(shop_A), 2)
mean_B = round(stat.mean(shop_B), 2)
med_A = round(stat.median(shop_A), 2)
med_B = round(stat.median(shop_B), 2)
if mean_B > med_B:
  print('ogish o\'ngga', mean_B, med_B)
else:
  print('ogish chapga', mean_B, med_B)
if mean_A > med_A:
  print('ogish o\'ngga', mean_A, med_A)
else:
  print('ogish chapga', mean_A, med_A)
range_A = round(max(shop_A) - min(shop_A), 1)
range_B = round(max(shop_B) - min(shop_B), 1)
print(f'Range for A: {range_A}',
      f'Range for B: {range_B}',
      f'Variance for A: {stat.pvariance(shop_A):.2f}',
      f'Variance for B: {stat.pvariance(shop_B):.2f}',
      f'Deviation for A: {stat.stdev(shop_A):.2f}',
      f'Deviation for B: {stat.stdev(shop_B):.2f}',
      f'A do\'konda barqarorlik balandroq, chunki, kunlik savdoda og\'ish pastroq {stat.stdev(shop_A):.2f}',
      sep = '\n'
      )
```

Berilgan ikta do'kon uchun savdo ko'rsatkichlari asosida olingan ma'lumotlarga asoslangan holda A do'kon berilgan 7 kunlik sotuv ko'rsatkichlarida sotuv miqdorlaridagi farqlarining tarqoqligi pasroq ya'ni sotuvlari barqarorroq.

### 3. 1000 qiymat natijaga qanday taâ€™sir qildi?

```python
# Shu ma'lumotlar bilan ishlang.

x = [1, 2, None, 4, "5", 6, "error"]
y = [10, 20, 30, None, "50", 60, 1000]
pairs = [
    (int(a), int(b))
    for a, b in zip(x, y)
    if isinstance(a, int)
       and isinstance(b, int)
       and b < 1000
]

x1 = [a for a, b in pairs]
y1 = [b for a, b in pairs]
print(
    f'Tozalangan x: {x1}',
    f'Tozalangan y: {y1}',
    f'Korrelyatsiya: {stat.correlation(x1, y1)}'
)
```

Bog'liq ma'lumotlarni chiqarib olganimizda ular orasida ideal korrelyatsiya mavjudligi aniqlanmoqda. Ya'ni har bir y ko'rsatkich x ko'rsatkichning 10 baravaridir
